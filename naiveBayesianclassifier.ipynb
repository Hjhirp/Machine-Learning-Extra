```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd

iris = datasets.load_iris()
X_train, X_test, Y_train, Y_test = train_test_split(iris.data[:,:], iris.target, test_size=0.3,random_state=109)
Y_train
```

Dividing My own Data set with values 0 , 1

```python
X = iris.data[:100, :2]
Y = iris.target[:100]
```

```python
X.size
```

```python
Y
```

```python
x_train=X[range(0,75,2),:]
y_train=Y[range(0,75,2)]
x_train.size
```

```python
x_test=X[range(1,75,2),:]
y_test=Y[range(1,75,2)]
x_test
```

```python
y_train.size
```

```python
plt.scatter(x_train[:,1:],y_train)
plt.grid(True)
```

```python
plt.scatter(x_train[:,:1],y_train)
plt.grid(True)
```

## **My data set calculation**

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(x_train, y_train).predict(x_test)
```

```python
print("Number of mislabeled points out of a total %d points : %d"
...       % (X_test.shape[0], (y_test != y_pred).sum()))
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
print(metrics.confusion_matrix(y_test, y_pred))
```

## **Actual** **calculation**

```python
gnb1 = GaussianNB()
y_pred1 = gnb.fit(X_train, Y_train).predict(X_test)
print(y_pred1)
print(Y_test)
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred1))
```
